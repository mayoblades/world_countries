```python
import pandas as pd
import json
import requests

url ='https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population'
def get_url(url):
    df = pd.read_html(url)
    df1 = df[0].iloc[0:,1:3]  # to take out the columns not needed
    df2 = pd.DataFrame(df1)
    df2.to_csv('countries.csv', index=False ,header=False)
    return df2
get_url(url)
```
